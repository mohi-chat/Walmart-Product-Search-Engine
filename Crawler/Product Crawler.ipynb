## Walmart Product Crawler

This ipython notebook contains implementation of web crawler to scrape product information from walmart.

<b> How it works? </b><br>
1). Provide a startlink (or seedlink) which is basically a link to a walmart product belonging to a category.<br>
2). Use a random surfer model to discover related products from product description.<br>
3). Stop when desired no of products have been found.
<br><br>

```python
import re
import time
import pandas as pd
import random
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
```

```python
#startlink = 'https://www.walmart.com/ip/Lenovo-81JW0001US-Chromebook-S330-14-HD-Display-Mediatek-MT8173C-CPU-4GB-RAM-32GB-eMMC-SSD-Chrome-OS-Black/206750547'
browser = webdriver.PhantomJS()
```

```python
startlink = 'https://www.walmart.com/ip/Xbox-Series-S/606518560'
```

```python
df = pd.DataFrame(columns=['sku_id', 'url', 'title', 'category', 'specifications', 'price', 'rating', 'no_of_reviews', 'outlink_sku_ids', 'outlinks'])
```

```python
index=0
crawled_set = set()
outlink_set = set()
```

```python
numSearch = 2223
while numSearch>0:
    try:
        browser.get(startlink)
        html = browser.page_source
        soup = BeautifulSoup(html, 'html.parser')
        skuId_element = soup.findAll("meta", {"itemprop": "sku"})
        if len(skuId_element) == 0:
            sample = random.sample(outlink_set, 1)[0]
            while sample[0] in crawled_set:
                sample = random.sample(outlink_set, 1)[0]
            startlink = sample[1]
            continue
        skuId = int(skuId_element[0]['content'])
        category = soup.find_all("nav", {"aria-label": "breadcrumb"})[0].text
        title = soup.findAll("h1", {"class": "prod-ProductTitle"})[0].get_text()
        price_holder = soup.findAll("span", {"id": "price"})
        if len(price_holder)>0:
            price_element = price_holder[0].findAll("span", {"class": "visuallyhidden"})
            if len(price_element)>0:
                price = price_element[0].get_text()
            else:
                price = 'N/A'
        else:
            price = 'N/A'
        rating, no_of_reviews = re.findall(r'\d+\.*\d*', soup.findAll("div", {"class":"prod-productsecondaryinformation"})[0].findAll("span", {"class": "stars-container"})[0]['aria-label'])
        specification_dict = {}
        table = soup.find('table', attrs={'class':'product-specification-table table-striped'})
        if table is not None:
            table_body = table.find('tbody')
            rows = table_body.find_all('tr')
            for row in rows:
                cols = row.find_all('td')
                specification_dict[cols[0].text] = cols[1].text
        outlink_skuids = []
        outlinks = []
        outlink_elements = soup.find_all("a", {"class": "tile-link-overlay"})
        for outlink in outlink_elements:
            if outlink['href'][1]!='i': continue
            parts = outlink['href'].split("/")
            outlink_skuids.append(parts[3].split('?')[0])
            outlinks.append('https://www.walmart.com'+'/'.join(parts[:3]) + '/' +parts[3].split("?")[0])
        df.loc[index] = [skuId, startlink, title, category, str(specification_dict), price, rating, no_of_reviews, str(outlink_skuids), str(outlinks)]
        crawled_set.add(skuId)
        outlink_set.update([(outlink_skuids[i],outlinks[i]) for i in range(len(outlinks))])
        sample = random.sample(outlink_set, 1)[0]
        while sample[0] in crawled_set:
            sample = random.sample(outlink_set, 1)[0]
        startlink = sample[1]
        index+=1
        numSearch-=1
        print("Done with "+str(index))
    except:
        print("Exception occured. Restarting browser after 1 Minute delay.")
        time.sleep(60)
        browser = webdriver.PhantomJS()
```

```python
df.to_csv('Video Games.csv', index=False)
```
